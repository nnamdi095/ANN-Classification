```
df=pd.read_csv('/content/drive/My Drive/Data Mining in Science/praktikum/3. Klasifikasi/iris_species.csv').dropna()
df.head()
```

# Data Preprocessing

```
X = df.iloc[:, :-1].values
y = df.iloc[:, 5].values
```

```
#Creating numeric classes for the dependent variable, y (Species)

from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
y
```

```
# Splitting the dataset into the Training set and Test set; working on 75% training set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Creating the ANN Model


```
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 3, init = 'uniform', activation = 'relu', input_dim = 5))

# Adding the second hidden layer
classifier.add(Dense(output_dim = 3, init = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 50)
```

The accuracy converged at 0.6339
Therefore, the ANN classification model is able to predict the out variable, y, with a 63.39% accuracy on the training set

# Making the predictions and evaluating the model - applying ANN model on the test set


```
# Predicting the Test set results
# Where the model is able to predict the output with a probability greater than 0.5, we make it true; otherwise, false

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

# Evaluating Model Prediction Accuracy

```
correct_prediction = 0

sample_space = y_pred.shape[0]

for result in y_pred:
  
    if result == True:
        correct_prediction += 1
        
accuracy = correct_prediction/sample_space*100

print("Total Sample: {}".format(sample_space))
print("\nCorrect Predictions: {}".format(correct_prediction))
print("\nAccuracy: {}".format(accuracy))
```

The ANN model is able to obtain a prediction accuracy of 65.79% on the test set
